```python
import time
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, layers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import utils
from BinaryConnect import create_model
from BinaryConnect import custom_train_cnn
import numpy as np

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Reshape and one-hot encode
x_train = x_train.reshape(-1, 784)  # Flatten the images
x_test = x_test.reshape(-1, 784)
y_train = utils.to_categorical(y_train, 10)  # One-hot encoding
y_test = utils.to_categorical(y_test, 10)
```

```python
# Define model and training parameters
epochs = 1000
batch_size = 200
validation_split = 0.1  # Use last 10000 samples of training set as validation set
early_stopping_patience = 80
weights_dict = {'deterministic': None, 'stochastic': None}

# Split data into training and validation sets
val_size = int(len(x_train) * validation_split)
x_val, y_val = x_train[-val_size:], y_train[-val_size:]
x_train, y_train = x_train[:-val_size], y_train[:-val_size]
```

```python
# Train models with both deterministic and stochastic binarization
for deterministic in [False, True]:
    test_errors = []
    train_times = []
    trained_epochs = []

    for i in range(6):#change the number of times the experiment is repeated
        # Create model
        model = create_model(deterministic)

        # Start timing model training
        start_time = time.time()

        # Train model using custom train function
        train_losses, val_accuracies = custom_train_cnn(model, x_train, y_train, x_val, y_val, epochs, batch_size, early_stopping_patience)


        # Calculate training time
        train_time = time.time() - start_time
        train_times.append(train_time)
        
        # Record the number of epochs actually trained
        trained_epochs.append(epochs)

        # # Evaluate model on test set
        test_loss, test_accuracy = model.evaluate(x_test, y_test)
        test_errors.append(1 - test_accuracy)

        # After training the model
        dense1_layer = model.layers[1]

        # Check if the layer has the attribute 'real_weights'
        if hasattr(dense1_layer, 'real_weights'):
            real_valued_weights = dense1_layer.real_weights.numpy()  # Get the real-valued weights
        else:
            # Fallback to the current weights if real_weights attribute doesn't exist
            real_valued_weights = dense1_layer.get_weights()[0]

        key = 'deterministic' if deterministic else 'stochastic'
        weights_dict[key] = real_valued_weights
    
    # Report average test error
    print(
        f"Deterministic Binarization: {deterministic}"
        f"\nAverage Test Error: {np.mean(test_errors)} +/- {np.std(test_errors)}"
        f"\nAverage Training Time: {np.mean(train_times)} seconds"
        f"\nAverage Number of Epochs: {np.mean(trained_epochs)}"
    )
```

```python
import matplotlib.pyplot as plt

#plot the histogram of real-valued weights in BinaryConnect MLP model in both methods, stochastically and deterministically

def plot_weight_histogram(weights_dict):
    plt.figure(figsize=(10, 5))

    for key, weights in weights_dict.items():
        # Flatten the weights to 1D for histogram
        weights_flat = weights.flatten()
        plt.hist(weights_flat, bins=300, alpha=0.5, label=f'{key} weights')

    plt.title('Weights Histogram for the first BinaryConnect layer')
    plt.xlabel('Weight values')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

# Call the function to plot the histogram
plot_weight_histogram(weights_dict)

```
